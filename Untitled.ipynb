```python
genres = sorted(set([g for genre_list in d_movie['genres'] for g in genre_list.split(',')]))
genre_matrices = []
for g in genres:
    genre_matrix = pd.DataFrame(columns=[''+g], index=d_movie.index, data=0)
    for i, row in d_movie.iterrows():
        if g in row['genres']:
            genre_matrix.loc[i, ''+g] = 1
    genre_matrices.append(genre_matrix)
```

```python
# Concatenate binary feature matrices horizontally
genre_matrix = pd.concat(genre_matrices, axis=1)
genre_matrix.values
```

```python
# Append binary feature matrix to original DataFrame
d_movie = pd.concat([d_movie, genre_matrix], axis=1)
d_movie.columns
```

```python
y = genre_matrix.values
```
